# **How is the correlation between cols?**

Correlation quantifies how two numbers change together. A positive correlation means they will increase or decrease together. A negative correlation means one increases as the other decreases. The closer the number is to 1 or -1, the stronger the linear relationship. Zero means no linear relationship.



````python
# How is the correlation between cols?
df.corr(numeric_only=True)
````

# **Label Based selection** (.loc)



```
# This is using .loc[ ], so it's a label-based selection.

.loc[row_labels, column_labels]

: in rows  select all rows

: in columns  select all columns

newdf = df.loc[:,:] It creates a full copy of the original DataFrame (df) into a new DataFrame called newdf
```



````python
newdf = df.loc[:,:]
newdf
````

````python
newdf.loc[0:3,'Survived':'Age']
````

# **Position-based indexing** (.iloc)



```
#  What is .iloc[]?
.iloc[] is for position-based indexing.

It uses integer positions, not label names.

What does 1:20, 1:4 mean?
1:20  Rows from position 1 up to 19 (not including 20)

1:4  Columns from position 1 up to 3 (not including 4)

So you're selecting:
Rows: 2nd to 20th row (index positions 1 to 19)

Columns: 2nd to 4th column
```



````python
df.iloc[1:20, 1:4]
````
