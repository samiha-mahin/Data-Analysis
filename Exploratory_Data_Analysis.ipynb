# **How is the correlation between cols?**

Correlation quantifies how two numbers change together. A positive correlation means they will increase or decrease together. A negative correlation means one increases as the other decreases. The closer the number is to 1 or -1, the stronger the linear relationship. Zero means no linear relationship.

# **Label Based selection** (.loc)



```
# This is using .loc[ ], so it's a label-based selection.

.loc[row_labels, column_labels]

: in rows → select all rows

: in columns → select all columns

newdf = df.loc[:,:] It creates a full copy of the original DataFrame (df) into a new DataFrame called newdf
```



````python
newdf = df.loc[:,:]
newdf
````

````python
newdf.loc[0:3,'Survived':'Age']
````

# **Position-based indexing** (.iloc)



```
#  What is .iloc[]?
.iloc[] is for position-based indexing.

It uses integer positions, not label names.

What does 1:20, 1:4 mean?
1:20 → Rows from position 1 up to 19 (not including 20)

1:4 → Columns from position 1 up to 3 (not including 4)

So you're selecting:
Rows: 2nd to 20th row (index positions 1 to 19)

Columns: 2nd to 4th column
```



````python
df.iloc[1:20, 1:4]
````

# **Exploratory Data Analysis(Bivariate & Multivariate Analysis)**

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
tips = sns.load_dataset('tips')
tips.head()
````

````python
titanic = pd.read_csv("titanic.csv")
titanic.head()
````

````python
flights = sns.load_dataset('flights')
flights.head()
````

````python
iris = sns.load_dataset('iris')
iris.head()
````

# **1.Scatter Plot (Numerical - Numerical)**

````python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue = tips['sex'],style = tips['smoker'])
````

# **2.Bar Plot (Numerical - Categorical)**

````python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
````

# **3.Box Plot (Numerical - Categorical)**

````python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
````

# **4.Distplot (Numerical - Categorical)**

````python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
````

# **5. Heatmap (Categorical - Categorical)**

````python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
````

````python
# Select the 'Survived' column after grouping by 'Pclass' and then calculate the mean
titanic.groupby('Pclass')['Survived'].mean() * 100
````

````python
flights.pivot_table(index='month',columns='year',values='passengers')
````

````python
sns.heatmap(flights.pivot_table(index='month',columns='year',values='passengers'))
````

# **6. Cluster Map (Categorical - Categorical)**

````python
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
````

````python
sns.clustermap(flights.pivot_table(index='month',columns='year',values='passengers'))
````

# **7.Pair Plot**

````python
sns.pairplot(iris,hue='species')
````
